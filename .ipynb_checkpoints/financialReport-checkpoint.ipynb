```python
# Import dependencies
import pandas as pd
import requests
from pymongo import MongoClient
import pymongo
from pprint import pprint
import csv
import json
from io import StringIO
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Connect to MongoDB server
client = pymongo.MongoClient('mongodb://localhost:27017/')
```

```python
# confirm that our new database was created
print(client.list_database_names())
```

```python
# assign the database to a variable name
df= client['Financial_Report']
```

```python
# Access the collection (replace 'data' with your collection name)
collection = df['data']
```

```python
# Fetch all documents from the collection
cursor = collection.find({})
```

```python
# Convert documents to list of dictionaries
documents = list(cursor)
```

```python
# review a document in the financial collection
pprint(df.data.find_one())
```

```python
# Convert list of dictionaries to pandas DataFrame
data_df = pd.DataFrame(documents)
# Display DataFrame
data_df
```

```python
#printing out the columns to reduce it
print(data_df.columns)
```

```python
#rewriting the column titles
columns_to_clean = [' Manufacturing Price ', ' Sale Price ', ' Gross Sales ', ' Discounts ', '  Sales ',' COGS ', ' Profit ']

# Loop through the columns to remove $ sign
for col in columns_to_clean:
    data_df[col] = pd.to_numeric(data_df[col].replace('[^0-9\.-]', '', regex=True), errors='coerce')
```

```python
# Rename columns for readability
updated_df = data_df.rename(columns={"_id": "id",
                                                  "Segment": "Segment",
                                                  "Country": "Country",
                                                  " Product ": "Product",
                                                  " Discount Band ": "Discount Band",
                                                  "Units Sold": "Units Sold",
                                                  " Manufacturing Price ": "Manufacturing Price",
                                                  "Sale Price": "Sale Price",
                                                  " Gross Sales ": "Gross Sales",
                                                  " Discounts ": "Discounts",
                                                  "  Sales ": "Sales",
                                                  " COGS ": "Cost of Goods Sold",
                                                  " Profit ": "Profit",
                                                  "Date": "Date",
                                                  "Month Number": "Month Number",
                                                  " Month Name ": "Month Name",
                                                  "Year": "Year"
                                                })

# Drop the rows where at least one element is missing.
updated_df = updated_df.dropna()
```

```python
# Reset the index to ensure continuous numbering
updated_df = updated_df.reset_index(drop=True)
```

```python
# Assuming 'columns_to_remove' 
reduce_columns = ["id", "Month Number"]

# Remove this specified columns
newupdated_df =updated_df.drop(columns=reduce_columns)

# Display the updated DataFrame
newupdated_df
```

```python
updated_df.describe()
```

```python
#list of countries in our dataset
updated_df['Country'].value_counts()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
counts = newupdated_df['Country'].value_counts()
counts.plot(kind="bar")
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Sales")
plt.show()
```

```python
# Convert float values to strings for the x-axis
x_axis = updated_df['Country'].astype(str)
y_axis = updated_df['Discounts']

# Create the bar plot
plt.bar(x_axis, y_axis, color='tab:blue', alpha=0.8, align='center')

# Set the xlabel and ylabel
plt.xlabel("Country")
plt.ylabel("Discounts")

# Rotate x-axis labels if needed
plt.xticks(rotation="vertical")

# Show the plot
plt.show()
```

```python

plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Discount Band', data=updated_df)

# Display count values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Count of Discount Band')
plt.xlabel('Discount Band')
plt.ylabel('Count')
plt.show()
```
